```python




T2D_vs_T2Dtreated = pd.read_csv('/content/T2D_vs_T2Dtreated.txt',sep="\t")
```

```python
T2D_vs_T2Dtreated = T2D_vs_T2Dtreated.drop(['ASVID','Kingdom','Phylum','Class','Order','Family',],axis=1)
T2D_vs_T2Dtreated = T2D_vs_T2Dtreated.set_index(['Genus'])
```

```python
IGT_vsIFG_IGT = IGT_vsIFG_IGT.drop(['ASVID','Kingdom','Phylum','Class','Order','Family',],axis=1)
IGT_vsIFG_IGT = IGT_vsIFG_IGT.set_index(['Genus'])
```

```python
IFG_IGT_vs_T2D = IFG_IGT_vs_T2D.drop(['ASVID','Kingdom','Phylum','Class','Order','Family',],axis=1)
IFG_IGT_vs_T2D = IFG_IGT_vs_T2D.set_index(['Genus'])
```

```python
healthy_vs_IFG = healthy_vs_IFG.drop(['ASVID','Kingdom','Phylum','Class','Order','Family',],axis=1)
healthy_vs_IFG = healthy_vs_IFG.set_index(['Genus'])
```

```python
IFG_vs_IGT = IFG_vs_IGT.drop(['ASVID','Kingdom','Phylum','Class','Order','Family',],axis=1)
IFG_vs_IGT = IFG_vs_IGT.set_index(['Genus'])
IFG_vs_IGT
```

```python
frames = [healthy_vs_IFG, IFG_vs_IGT,IGT_vsIFG_IGT,IFG_IGT_vs_T2D,T2D_vs_T2Dtreated]

result = pd.concat(frames,axis=1)
result
```
